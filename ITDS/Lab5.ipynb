```python
def gaussian(x, mu, sigma):
	return np.exp( -(x-mu)**2/(2*(sigma**2)) ) / (sigma * math.sqrt(2*math.pi))
```

```python
x = np.arange(-3,3,0.1)
mu = 0 # Mean
sigma = 1 # Standard deviation
plt.plot( x, gaussian(x, mu, sigma), label = r"$\mu$ = {}, $\sigma$ = {}".format(mu, sigma) )
plt.show()
plt.close()
```

```python
def flip_coin():
	if np.random.rand() < 0.5:
		return 0
	return 1

x = []
m = []
for N in range(10,1000,10):
    data = [flip_coin() for i in range(N)]
    x.append(N)
    m.append( np.mean(data) )

plt.scatter(x, m)
```

```python
def experiment(n, repetitions):
    means = []
    for i in range(repetitions):
        results = [flip_coin() for i in range(n)]
        mean = np.mean(results)
        means.append(mean)

    print("Mean deviation", np.std(means)/ math.sqrt(n))
    plt.hist(means, bins=25)

experiment(100, 1000)
```
