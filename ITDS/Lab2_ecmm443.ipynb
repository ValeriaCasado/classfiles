```python
import numpy as np
import matplotlib.pyplot as plt
%inline matplotlib
```

```python
f = open("salaries.txt", 'r')

X = [float(line[:-3]) for line in f.readlines()]
Y = np.ones_like(X)
```

```python
print(X[:10])
```

```python
fig, ax = plt.subplots()
ax.scatter(X,Y, alpha=0.1)
plt.show()
```

## Excercise 2
Binning
A useful way to examine our data is to ask questions like "What
proportion of jobs have salaries in the range £25000 to £30000?".
Choosing an increment and counting up all the data points that
lie in each interval is called binning and the plot of the counts in
each bin is called a histogram. Matplotlib has a function which
constructs histograms from lists of data points.
plt.hist(data, bins = 100)
plt.show()

```python
plt.hist(X, bins=100)
plt.show()
```

```python
above100k = [x for x in X if x >=100000]
```

```python
plt.hist(above100k, bins=50)
plt.show()
```

```python
bins = [5000*i for i in range(21)] #Shows bins sub 100k
print(bins)
plt.hist(X, bins=bins)
plt.show()
```

```python
plt.hist(X, bins = bins, label = "Salaries")
plt.xlabel("GBP")
plt.ylabel("Number of Jobs")
plt.legend()
plt.title("Histogram of Salary Data")
plt.savefig("salary_histogram.png")
```

```python
plt.hist(X, bins = bins, label = "Salaries")
plt.xlim(0,100000)
plt.xlabel("GBP")
plt.ylabel("Number of Jobs")
plt.legend(loc="lower right")
plt.title("Histogram of Salary Data")
```

```python
plt.hist(X, bins=bins, cumulative=True, histtype='step', density=True)
plt.show()
```

```python
plt.hist(X, bins=bins, cumulative=False, histtype='step', density=True)
plt.show()
```
