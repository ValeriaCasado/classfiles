# Exercise 1

"But that is the beginning of a new story the story of the gradual renewal of a man the story of his gradual regeneration of his passing from one world into another of his initiation into a new unknown life that might be the subject of a new story but our present story is ended"

Write python code (using any libraries you like) to count the frequency of each word in the text above and plot a bar chart showing the words that appear more than once. 



```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import cm


string = "But that is the beginning of a new story the story of the gradual renewal of a man the story of his gradual regeneration of his passing from one world into another of his initiation into a new unknown life that might be the subject of a new story but our present story is ended"

str_arr = list(set(string.split(" ")))
d = dict.fromkeys(str_arr,0)

for word in string.split(" "):
    d[word]+=1

print(str_arr)
fig, ax = plt.subplots()
fig.set_size_inches(10.5, 4.5)

ax.bar(list(d.keys()), list(d.values()), color="#39ff00")
ax.tick_params(axis='x', labelsize=12, rotation=290)
ax.set_xlabel("-- Words --")
ax.xaxis.get_label().set_fontsize(16)
ax.set_ylabel("-- Count --")
ax.yaxis.get_label().set_fontsize(16)
#ax.set_facecolor("#d8d8d8")
#ax.legend(frameon=False)
ax.grid(color="#d8d8d8", linewidth=1)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(False)
ax.set_axisbelow(True)

```

```python
from scipy.stats import norm
import math
import numpy as np
def gaussian(x, mu, sigma):
	return np.exp( -(x-mu)**2/(2*(sigma**2)) ) / (sigma * math.sqrt(2*math.pi))

mu = 0 # Mean
sigma = 1 # Standard deviation

dataset=[1,3,-2,3]

def nthmoment(n, dataset):
    moment = 0
    for value in dataset:
        moment = (value**n)*gaussian(value, mu, sigma)
    return moment

nthmoment(2,dataset)

```
